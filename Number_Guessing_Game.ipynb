```python
print("Welcome to the Guessing Game!, Pick a number 1-100")
import random
answer= random.randint(1,100)
guessed=False
num_guesses = 0

while not guessed:
  try:
    guess = int(input("Guess a number between 0 and 100: "))
  except ValueError:
    print("Invalid input. Please enter a whole number.")
    continue

  num_guesses += 1

  if guess == answer:
    guessed = True
    print("You got it! The number was", answer, "You guessed it in", num_guesses, "guesses!")
  elif guess > answer:
    print("Too high! Guess lower.")
  else:
    print("Too low! Guess higher.")

print("Thanks for playing!")


```
