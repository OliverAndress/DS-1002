```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO

def red_filter(photo):
    filtered_photo = np.copy(photo)
    filtered_photo[:, :, 1:] = 0  # Set G and B values to zero
    return filtered_photo

def green_filter(photo):
    filtered_photo = np.copy(photo)
    filtered_photo[:, :, 0] = 0  # Set R values to zero
    filtered_photo[:, :, 2] = 0  # Set B values to zero
    return filtered_photo

def blue_filter(photo):
    filtered_photo = np.copy(photo)
    filtered_photo[:, :, :2] = 0  # Set R and G values to zero
    return filtered_photo

def main():
    url = input("Enter the URL of the .jpg file: ")
    response = requests.get(url)
    original_image = Image.open(BytesIO(response.content))
    original_photo = np.array(original_image)

    filtered_images = {
        'Red Filtered': red_filter(original_photo),
        'Green Filtered': green_filter(original_photo),
        'Blue Filtered': blue_filter(original_photo),
        'No Blue': original_photo[:, :, :2],  # R and G values but no blue
        'No Green': np.concatenate((original_photo[:, :, 0:1], np.zeros_like(original_photo[:, :, 1:2]), original_photo[:, :, 2:]), axis=2),  # R and B values but no green
        'No Red': np.concatenate((np.zeros_like(original_photo[:, :, :1]), original_photo[:, :, 1:]), axis=2)  # G and B values but no red
    }

    print("Original Image:")
    display(original_image)

    for label, photo in filtered_images.items():
        filtered_image = Image.fromarray(photo.astype(np.uint8))
        print(label + ":")
        display(filtered_image)

    # Check if the original image is identical to the combination of filtered images
    combined_image = np.stack((filtered_images['Red Filtered'][:, :, 0], filtered_images['Green Filtered'][:, :, 1], filtered_images['Blue Filtered'][:, :, 2]), axis=2)
    if np.array_equal(original_photo, combined_image):
        print("Original image is identical to the combination of filtered images.")
    else:
        print("Original image is not identical to the combination of filtered images.")

if __name__ == "__main__":
    main()
```
