```python
#Bonus
from collections import Counter


letter_pair_counts = Counter() # Counts occurrences of each letter pair
for word in wordlist:
    word_lower = word.lower()
    for i in range(len(word_lower) - 1):
        letter_pair_counts[word_lower[i:i+2]] += 1  # Counts occurrences of consecutive pairs


most_common_pair = letter_pair_counts.most_common(1)[0][0]
least_common_pair = letter_pair_counts.most_common()[-1][0]


most_common_count = letter_pair_counts[most_common_pair]
least_common_count = letter_pair_counts[least_common_pair] # Counts occurrences of the most and least common pairs

print("Pair of letters in most words:", most_common_pair, most_common_count, "words")
print("Pair of letters in fewest words:", least_common_pair, least_common_count, "words")
```
